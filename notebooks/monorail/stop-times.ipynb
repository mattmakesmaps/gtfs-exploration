## Stop Times

```python
## Get a dataframe going.

import os
import pandas as pd

data_dir = os.path.join(os.getcwd(), os.pardir, os.pardir, 'data')
stop_times_csv =  os.path.join(data_dir, 'monorail', 'stop_times.txt')
stop_times_df = pd.read_csv(stop_times_csv)

stop_times_df.head()
```

Create a table showing the latest departure times for regular schedule and special events.

```python
## Remove the timestamp from `trip_id`
trip_type = stop_times_df["trip_id"].str.rpartition("_")[0]
stop_times_df = stop_times_df.assign(trip_type=trip_type)

# Perform SELECT, GROUPBY, and MAX aggregation
latest_trips_df = stop_times_df[["trip_type", "departure_time"]].groupby("trip_type").agg({"departure_time": "max"})
latest_trips_df.sort_values("trip_type")
```
