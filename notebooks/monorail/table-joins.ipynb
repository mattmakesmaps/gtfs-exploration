## Joining Data Frames

```python
## Get a dataframe going.

import os
import pandas as pd

data_dir = os.path.join(os.getcwd(), os.pardir, os.pardir, 'data')
stop_times_csv =  os.path.join(data_dir, 'monorail', 'stop_times.txt')
stops_csv =  os.path.join(data_dir, 'monorail', 'stops.txt')

stop_times_df = pd.read_csv(stop_times_csv)
stops_df = pd.read_csv(stops_csv)
```

```python
stop_times_df
```

```python
stops_df
```

```python
## Perform an inner join

joined_stops_df = pd.merge(stop_times_df, stops_df, on='stop_id')
joined_stops_df
```
